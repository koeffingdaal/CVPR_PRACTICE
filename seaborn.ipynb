```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# To display plots inside notebook
%matplotlib inline
```

```python
dataset = sns.get_dataset_names()
dataset
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
attention = sns.load_dataset("attention")
attention
```

```python
sns.lineplot(x = "total_bill", y = "tip", data = tips)
plt.title("Lineplot based on TIPS");
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.lineplot(x = "sepal_length", y = "sepal_width", data = iris)
plt.title("iris dataset");
```

```python
sns.stripplot(y = 'sepal_length', x = 'species', data = iris)
plt.title("strripplot");
```

```python
sns.stripplot(x='day', y='total_bill', data=tips,
             jitter=True,
             hue='smoker',
             dodge=True
             )
plt.title("Stripplot based on 'tips' dataset")
```

```python
sns.swarmplot(y = 'sepal_length', x = 'species', data = iris)
plt.title("strripplot");
```

```python
sns.barplot(y="sepal_length", x="species", data=iris, palette = "plasma")
plt.title("Barplot based on 'iris' dataset");
```

```python
sns.barplot(y="sepal_length", x="species", data=iris)
plt.title("Barplot based on 'iris' dataset");
```
